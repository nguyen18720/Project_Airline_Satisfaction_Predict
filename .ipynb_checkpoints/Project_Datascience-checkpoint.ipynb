## 2. Visualizing

### Customer's properties

- Select customer's properties columns

```python
bin_user_prop = [i for i in df.loc[:,"satisfaction":"Class"].columns if i != "Age" ]
```

```python
df[bin_user_prop]
```

- Using countplot to visualize customer's properties along with satisfaction

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8), dpi= 200)
for i in range(0,2):
    for j in range(0,2):
        sns.countplot(data= df[bin_user_prop], x= "satisfaction", hue= bin_user_prop[i+j+1], ax= axes[i][j])
        axes[i][j].legend(loc= (1.01, 0.87))
        axes[i][j].set_xlabel("")
        axes[i][j].set_ylabel("")
        axes[i][j].set(title= bin_user_prop[i+j+1])

plt.tight_layout()
plt.show()
```

- Separate customers into 4 groups according to Age  

```python
df["Age_cat"]= pd.cut(df["Age"], bins= [7, 27, 57, 77, np.inf]
        , labels= ["7-27", "27-57", "57-77", "77-85"] )
```

```python
plt.figure(figsize= (12, 5), dpi= 200)
sns.countplot(data= df, x= "Age_cat", hue= "satisfaction");
```
